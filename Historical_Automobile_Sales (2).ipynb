```python
sales_by_vehicle = df.groupby(['Recession', 'Vehicle_Type'], as_index=False)['Automobile_Sales'].mean()

plt.figure(figsize=(10,6))
sns.barplot(data=sales_by_vehicle, x='Vehicle_Type', y='Automobile_Sales', hue='Recession')
plt.title('Average Vehicle Sales During Recession and Non-Recession')
plt.xlabel('Vehicle Type')
plt.ylabel('Average Automobile Sales')
plt.legend(title='Recession', labels=['Non-Recession', 'Recession'])
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
recession_data = df[df['Recession'] == 1]
non_recession_data = df[df['Recession'] == 0]

gdp_recession = recession_data.groupby('Year', as_index=False)['GDP'].mean()
gdp_non_recession = non_recession_data.groupby('Year', as_index=False)['GDP'].mean()

fig, axes = plt.subplots(1, 2, figsize=(14,5), sharey=True)

axes[0].plot(gdp_recession['Year'], gdp_recession['GDP'], marker='o')
axes[0].set_title('GDP During Recession')
axes[0].set_xlabel('Year')
axes[0].set_ylabel('Average GDP')
axes[0].tick_params(axis='x', rotation=45)

axes[1].plot(gdp_non_recession['Year'], gdp_non_recession['GDP'], marker='o')
axes[1].set_title('GDP During Non-Recession')
axes[1].set_xlabel('Year')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
seasonality_data = df.groupby('Month', as_index=False)[['Automobile_Sales', 'Seasonality_Weight']].mean()

plt.figure(figsize=(10,6))
plt.scatter(
    seasonality_data['Month'],
    seasonality_data['Automobile_Sales'],
    s=seasonality_data['Seasonality_Weight'] * 2000,
    alpha=0.6
)
plt.title('Impact of Seasonality on Automobile Sales')
plt.xlabel('Month')
plt.ylabel('Average Automobile Sales')
plt.tight_layout()
plt.show()
```

```python
recession_data = df[df['Recession'] == 1]

plt.figure(figsize=(10,6))
sns.scatterplot(data=recession_data, x='Price', y='Automobile_Sales', hue='Vehicle_Type')
plt.title('Vehicle Price vs Automobile Sales During Recession')
plt.xlabel('Average Vehicle Price')
plt.ylabel('Automobile Sales')
plt.tight_layout()
plt.show()
```

```python
ad_exp = df.groupby('Recession', as_index=False)['Advertising_Expenditure'].sum()

plt.figure(figsize=(8,8))
plt.pie(
    ad_exp['Advertising_Expenditure'],
    labels=['Non-Recession', 'Recession'],
    autopct='%1.1f%%',
    startangle=90
)
plt.title('Advertising Expenditure During Recession and Non-Recession')
plt.tight_layout()
plt.show()
```

```python
recession_data = df[df['Recession'] == 1]
ad_by_vehicle = recession_data.groupby('Vehicle_Type', as_index=False)['Advertising_Expenditure'].sum()

plt.figure(figsize=(8,8))
plt.pie(
    ad_by_vehicle['Advertising_Expenditure'],
    labels=ad_by_vehicle['Vehicle_Type'],
    autopct='%1.1f%%',
    startangle=90
)
plt.title('Advertisement Expenditure by Vehicle Type During Recession')
plt.tight_layout()
plt.show()
```
